Paso 1 - Importar librerías

```python
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)
```

Paso 2 - Cargar el dataset

```python
from google.colab import files

uploaded = files.upload()
```

```python
df = pd.read_json("streaming_users_dirty.json")
```

Paso 3 - Mostrar información genera

```python
df.info()
```

Paso 4 - Ver los primeros registros

```python
df.head()
```

Paso 5 - Buscar valores nulos

```python
df.isnull().sum()
```

Paso 6 - Completar valores faltantes

```python
for columna in df.select_dtypes(include=["int64","float64"]).columns:
    df[columna] = df[columna].fillna(df[columna].median())
```

```python
for columna in df.select_dtypes(include="object").columns:
    df[columna] = df[columna].fillna(df[columna].mode()[0])
```

Paso 7 - Verificar que ya no existan nulos

```python
df.isnull().sum()
```

Paso 8 - Buscar registros duplicados

```python
df.duplicated().sum()
```

Paso 9 - Eliminar registros duplicados

```python
df = df.drop_duplicates()
```

Paso 10 - Verificar nuevamente

```python
df.duplicated().sum()
```

Paso 11 - Revisar categorías

```python
for columna in df.select_dtypes(include="object").columns:
    print("\n", columna)
    print(df[columna].value_counts())
```

Paso 12 - Corregir planes

```python
df["subscription_plan"] = df["subscription_plan"].replace({
    "basico": "Básico",
    "BASICO": "Básico",
    "Basic": "Básico",
    "básico": "Básico",

    "Std": "Estándar",
    "estandar": "Estándar",
    "STANDARD": "Estándar",
    "Estándar ": "Estándar",
    "Standard": "Estándar",

    "Premium ": "Premium",
    "PREMIUM": "Premium",
    "Premiun": "Premium",
    "premium": "Premium"
})
```

```python
df["subscription_plan"].value_counts()
```

Paso 13 - Corregir países

```python
df["country"] = df["country"].str.strip()

df["country"] = df["country"].replace({

    "colombia": "Colombia",
    "COL": "Colombia",

    "méxico": "México",
    "Mexico": "México",
    "MEX": "México",

    "uruguay": "Uruguay",
    "URY": "Uruguay",

    "Brazil": "Brasil",
    "BRA": "Brasil",
    "brasil": "Brasil",

    "Peru": "Perú",
    "PER": "Perú",
    "perú": "Perú",

    "CHL": "Chile",
    "chile": "Chile",

    "argentina": "Argentina",
    "ARG": "Argentina"
})
```

```python
df["country"].value_counts()
```

Paso 14 - Corregir géneros

```python
df["favorite_genre"] = df["favorite_genre"].str.strip()

df["favorite_genre"] = df["favorite_genre"].replace({

    "Action": "Acción",
    "ACCIÓN": "Acción",
    "accion": "Acción",

    "Comedy": "Comedia",
    "COMEDIA": "Comedia",
    "comedy": "Comedia",
    "Comedia ": "Comedia",

    "Crime": "Crimen",
    "CRIME": "Crimen",
    "crime": "Crimen",

    "Documentary": "Documental",
    "DOC": "Documental",
    "documental": "Documental",

    "Drama": "Drama",
    "DRAMA": "Drama",
    "drama": "Drama",

    "ROMANCE": "Romance",
    "romance": "Romance",
    "Romance ": "Romance",

    "THRILLER": "Thriller",
    "Thriller ": "Thriller",
    "thriler": "Thriller"
})
```

```python
df["favorite_genre"].value_counts()
```

Paso 15 - Convertir fechas

```python
df["last_login_date"] = df["last_login_date"].replace({

    "0000-00-00": pd.NA,
    "not_available": pd.NA,
    "2026-15-03": pd.NA,
    "31-02-2022": pd.NA,
    "2029-01-01": pd.NA

})
```

```python
df["last_login_date"] = pd.to_datetime(
    df["last_login_date"],
    errors="coerce"
)
```

```python
df["last_login_date"] = df["last_login_date"].fillna(
    df["last_login_date"].mode()[0]
)
```

Paso 16 - Pasarlas al formato AAAA-MM-DD (2000-10-20)

```python
df["last_login_date"] = df["last_login_date"].dt.strftime("%Y-%m-%d")
```

Paso 17 - Revisar edades

```python
df["age"].describe()
```

Paso 18 - Corregir edades fuera de rango



```python
mediana = df["age"].median()

df.loc[df["age"] < 13, "age"] = mediana
df.loc[df["age"] > 100, "age"] = mediana
```

Paso 19 - Revisar tiempo de visualización

```python
df["monthly_watch_time_mins"].describe()
```

Paso 20 - Corregir tiempos negativos

```python
mediana = df["monthly_watch_time_mins"].median()

df.loc[df["monthly_watch_time_mins"] < 0,
       "monthly_watch_time_mins"] = mediana
```

Paso 21 - Revisar tickets de soporte

```python
df["customer_support_tickets"].describe()
```

Paso 22 - Verificación final

```python
print(df.isnull().sum())

print(df.duplicated().sum())

print(df.shape)
```

Paso 23 - Guardar el dataset limpio

```python
df.to_json(
    "streaming_users_clean.json",
    orient="records",
    indent=4
)
```
